## Procedimentos Aninhados

```python
def pessoa(a):
    return a + " vai a/ao"
```

```python
def lugar(b):
    return b + "!!!"
```

```python
def frase(a,b):
    sujeito = pessoa(a)
    local = lugar(b)
    return sujeito + " " + local
    
```

```python
frase('João','Angra')
```

Podemos aninhar procedimentos e essa é uma as funções do retorno, ou seja, passar um parâmetro para outro procedimento e obter assim um resultado mais complexo 

```python
#Dados 3 números, descubra o maior deles.
#Há varias formas de resolver o problema
```

```python
def maiorNum(a,b):
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return a
```

```python
maiorNum(21,20)
```

```python
def maiorTres(a,b,c):
    aux = maiorNum(a,b)
    if aux > c:
        return aux
    elif c > aux:
        return c
    else:
        return aux
```

```python
maiorTres(56,653,862)
```
